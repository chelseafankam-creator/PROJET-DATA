```python
import pandas as pd
from IPython.display import display

df = pd.read_csv("animes.csv")
df=df.drop_duplicates()
df["Note_Globale"]=pd.to_numeric(df["Note_Globale"],errors="coerce")
df["Note_Meilleur_Ep"]=pd.to_numeric(df["Note_Meilleur_Ep"],errors="coerce")
df["Note_Pire_Ep"]=pd.to_numeric(df["Note_Pire_Ep"],errors="coerce")
df=df.dropna(subset=["Note_Globale", "Note_Meilleur_Ep", "Note_Pire_Ep"])
df=df.sort_values(by="Anime", ascending=True)
df=df.reset_index(drop=True)
print("---Types de données---")
display(df.dtypes)

print("\n--- Aperçu du dataset nettoyé ---")
display(df.head())

df.to_csv("animes_propre.csv", index=False)
print("\nFichier 'animes_propre.csv' créé avec succès !")
print(f"Nombre Total d'animés restants {len(df)}")
```

```python
corr=df["Note_Globale"].corr(df["Note_Meilleur_Ep"])
print(f" correlation Note Globale vs Note Meilleur Ep:{corr:.2f}")
if corr> 0.5:
    print("\ncorrelation POSITIVE forte")
    print("Une excellente Episode booste la Note Globale")
elif corr>0:
    print("\n correlation POSITIVE faible")
else:
    print("\ncorrelation NEGATIVE")
```

```python
corr=df["Note_Globale"].corr(df["Note_Pire_Ep"])
print(f" correlation Note Globale vs Note Pire Ep:{corr:.2f}")
print(f"\nRECAP:")
print(f"\nNote Meilleur Ep:correlation POSITIVE forte")
print(f" Note Pire Ep:correlation plus faible")
print(f"un anime peut etre nul en episode mais reste bien noté")
```

```python
df["Ecart"]=df["Note_Meilleur_Ep"]-df["Note_Pire_Ep"]
df["Regularite"]=10 - df["Ecart"]
df["Score_Metier"]=(df["Note_Globale"] * 0.7) + (df["Regularite"] * 0.3) 
moyenne_metier=df["Score_Metier"].mean()
print(f"La Moyenne c'est : {moyenne_metier:.2f}/10")
score_max=df["Score_Metier"].max()
meilleur_anime=df[df["Score_Metier"]==score_max]["Anime"].values[0]
print(f"l'anime numero 1 est: {meilleur_anime}")
display(df[["Anime","Note_Globale","Regularite","Score_Metier"]].head(10))
```

```python
df["Date_Pub"]=pd.to_datetime(df["Date_Pub"])
df["Annee"]=df["Date_Pub"].dt.year
print(f"LES DATES ONT ETE CONVERTIES ET LES ANNÉES EXTRAITE:")
print("-" * 40)
print(df[['Anime', 'Date_Pub', 'Annee']].head(10))
```
